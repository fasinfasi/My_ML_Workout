```python
from sklearn.datasets import load_digits
import pandas as pd
digit = load_digits()
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB, MultinomialNB
```

```python
model_para = {
    'svm' : {
        'model' : SVC(gamma='auto'), 
        'para' : {
            'kernel' : ['linear', 'rbf'], 
            'C' : [1 ,10 ,20]
        }
    }, 
    'random_forest' : {
        'model' : RandomForestClassifier(),
        'para' : {
            'n_estimators' : [1, 5, 10]
        }
    }, 
    'naive_bayes_gausssian' : {
        'model' : GaussianNB(), 
        'para' : {}
    }, 
    'naive_bayes_multinomial' : {
        'model' : GaussianNB(), 
        'para' : {}
    }, 
    'logisctic_regression' : {
        'model' : LogisticRegression(), 
        'para' : {
            'C' : [1, 5, 10]
        }
    }, 
    'decision_tree' : {
        'model' : DecisionTreeClassifier(), 
        'para' : {
            'criterion' : ['gini', 'entropy']
        }
    }
}
```

```python
model_para.items()
```

#### GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
scores = []

for model_name, mp in model_para.items():
    clf =  GridSearchCV(mp['model'], mp['para'], cv=5, return_train_score=False)
    clf.fit(digit.data, digit.target)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })

df = pd.DataFrame(scores, columns=['model', 'best_score', 'best_para'])
df
```

#### For me the winner is svm (C=1, kernel=linear) with 94.76% score. It could be different for you as I have limited my parameters to be certain values only
